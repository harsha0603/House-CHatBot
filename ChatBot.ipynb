```python
def query_houses(dataframe, location = None, max_price = None, min_sqft = None):
    filtered_df = dataframe.copy()
    
    if location:
        filtered_df = filtered_df[filtered_df['location'].str.contains(location, case = False)]
        
        if max_price is not None:
            filtered_df = filtered_df[filtered_df['price'] <= max_price]
            
        if min_sqft is not None:
            filtered_df = filtered_df[filtered_df['total_sqft'] >= min_sqft]
            
        return filtered_df
    
results = query_houses(df, location="Whitefield",max_price=231.00, min_sqft=1521)
print(results.head())
```

```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")

def extract_parameters(user_input):
    # Process the original input for entity recognition
    doc = nlp(user_input)

    location = None
    max_price = None
    min_sqft = None

    # Loop over detected entities
    for ent in doc.ents:
        if ent.label_ == "GPE":
            if location is None:
                location = ent.text
        elif ent.label_ in ["MONEY", "CARDINAL"]:
            text_lower = ent.text.lower()
            if "lakh" in text_lower or "crore" in text_lower:
                # Remove non-digit characters for conversion
                price_text = re.sub(r'[^\d.]', '', text_lower)
                try:
                    if "lakh" in text_lower:
                        # Convert lakhs to thousands: 1 lakh = 100 thousands
                        max_price = float(price_text) * 100  
                    elif "crore" in text_lower:
                        # Convert crores to thousands: 1 crore = 10000 thousands
                        max_price = float(price_text) * 10000
                except ValueError:
                    pass
            else:
                # Treat as thousands if no unit is specified
                try:
                    max_price = float(ent.text)
                except ValueError:
                    pass

    # Use regex for sqft extraction on lowercased text
    match = re.search(r'(\d{3,5})\s*(sqft|square feet|sq ft)?', user_input.lower())
    if match:
        min_sqft = int(match.group(1))

    return {"location": location, "max_price": max_price, "min_sqft": min_sqft}

# Test with a sample query
user_query = "Show me houses in Whitefield under 231 lakhs at least 1521 sqft"
params = extract_parameters(user_query)
print(params)
```

```python
df_scaled['price'] = pd.to_numeric(df_scaled['price'], errors='coerce')
df_scaled['total_sqft'] = pd.to_numeric(df_scaled['total_sqft'], errors='coerce')
```

```python
df_scaled.dropna(subset= features, inplace=True)
```

```python
df_scaled[features] = scalar.fit_transform(df_scaled[features])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_similar_houses_with_location(filtered_df, df_scaled, target_index, top_n=5):
    
    # Extract the feature matrix from the scaled DataFrame for houses in the filtered subset.
    feature_matrix = df_scaled.loc[filtered_df.index, ['price', 'total_sqft']].values
    
    # Compute cosine similarity among these houses.
    similarity_matrix = cosine_similarity(feature_matrix)
    
    # Get similarity scores for the target house.
    target_similarity = similarity_matrix[target_index]
    
    # Get indices for top similar houses, excluding the target itself.
    similar_indices = np.argsort(-target_similarity)[1:top_n+1]
    
    # Map these indices back to the filtered DataFrame.
    similar_houses = filtered_df.iloc[similar_indices]
    
    return similar_houses

# Example usage:
# Filter the DataFrame by location (and other criteria if needed)
filtered = query_houses(df, location="Whitefield", max_price=231, min_sqft=1521)

if not filtered.empty:
    # Let's choose the first house in the filtered results as our target.
    similar = get_similar_houses_with_location(filtered, df_scaled, target_index=0, top_n=5)
    print("Top similar houses in the selected location:")
    print(similar[['location', 'price', 'total_sqft']])
else:
    print("No houses found in the specified location with the given criteria.")
```

```python
def generate_response(filtered_df, similar_houses, location, max_price, min_sqft):
    if filtered_df.empty:
        return f"Sorry, we couldn't find any houses in {location} under {max_price} (in lakhs) with at least {min_sqft} sqft."
    
    response = f"Here are some recommended houses in {location} (with price up to {max_price} and size at least {min_sqft} sqft):\n\n"
    for idx, row in similar_houses.iterrows():
        response += (f"- {row['area_type']} in {row['location']} has {row['total_sqft']} sqft, "
                     f"priced at {row['price']} (in lakhs), and is available {row['availability']}.\n")
    return response


def chatbot():
    user_query = input("Enter your query: ")  # e.g., "Show me houses in Whitefield under 231 at least 1521 sqft"
    
    
    params = extract_parameters(user_query)
    location = params.get("location")
    max_price = params.get("max_price")
    min_sqft = params.get("min_sqft")
    
   
    filtered = query_houses(df, location=location, max_price=max_price, min_sqft=min_sqft)
    
    if filtered.empty:
        print(f"Sorry, we couldn't find any houses in {location} under {max_price} with at least {min_sqft} sqft.")
        return
    
   
    target_index = 0
    similar = get_similar_houses_with_location(filtered, df_scaled, target_index, top_n=5)
    
    
    response = generate_response(filtered, similar, location, max_price, min_sqft)
    print("\n" + response)


if __name__ == "__main__":
    chatbot()

```
